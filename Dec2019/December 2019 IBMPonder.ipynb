```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x_u(u):
    x = (1.0-2.0*u)/(99998.0*(u-1.0)*u-1.0)
    return x
```

```python
def y_u(u):
    y = -2.0000400008e-5*((u**2-u+0.5)/(u**2-u-1.0000200004e-5))
    return y
```

```python
u = np.linspace(0.0,1.0,1000)
x = np.zeros(1000)
y = np.zeros(1000)
absolute_value = np.zeros(1000)
for n in range(len(u)):
    x[n] = x_u(u[n])
    y[n] = y_u(u[n])
    absolute_value[n] = np.abs(x[n])
```

```python
# Mean square error
np.square(np.subtract(y,absolute_value)).mean()
```

```python
plt.figure(figsize=(7.0,7.0/((np.sqrt(5)+1)/2)))
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12
plt.plot(x,absolute_value,c='red',label='|x|')
plt.plot(x,y,'-.',c='blue',label='Approximation')
plt.legend()
plt.show()
```
