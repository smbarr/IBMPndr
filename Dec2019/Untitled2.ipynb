```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,1000)
freq = np.fft.fftfreq(1000)
f = np.abs(x)
fft = np.fft.fft(f)
```

```python
plt.plot(freq[0:100],np.real(fft)[0:100])
plt.plot(freq[901:],np.real(fft)[901:])
#plt.xlim(-0.01,0.01)
```

```python
plt.plot(freq,np.imag(fft))
plt.xlim(-0.01,0.01)
```
